```python
# Lisää vastauksesi ainoastaan python funktion sisälle, ennen print komentoa:
def suomisahko(url):
    import pandas as pd
    import numpy as np
        
    df = pd.read_csv(url,delimiter=",",decimal=".")
    df[df.loc[:,"January €/MWh":"December €/MWh":1].isin(["."])]=np.nan
        
    df['Year Total €/MWh'].replace(".",np.nan,inplace=True)    
    df.loc[:,"Year Total €/MWh":"December €/MWh":1] = df.loc[:,"Year Total €/MWh":"December €/MWh":1].astype(float)
    
    df=df.loc[df['Price area'] =='Finland']    
    df['Average €/MWh']=df.loc[:,"January €/MWh":"December €/MWh":1].mean(axis='columns')
    
    df['Below average price']=(df.loc[:,"January €/MWh":"Average €/MWh":1]).apply(lambda x: (x[0:12] < x[12]).sum(), axis=1)    
    df['Above average price']=(df.loc[:,"January €/MWh":"Average €/MWh":1]).apply(lambda x: (x[0:12] >= x[12]).sum(), axis=1)
    
    df=df[(df['Below average price']+df['Above average price'])==12]    
    df=df[['Year','Below average price','Above average price']]
    
    tulos=df.head(10)    
    print(tulos)        
    
url="https://student.labranet.jamk.fi/~huoptu/datananal_s20/kerta2/harjoitukset/ehi_010_2020q2.csv"
suomisahko(url)     
```
