## Datan visualisointi  (T6.3 / 1p)

Toteuta seuraavat kohdat tässä tehtävässä:

Visualisoi aikasarjadata haluamallasi tavalla.
* Valitse tähän ainakin kaksi eri kuvaajatyyppiä
* Visualisoi data kenttien 'Low price' ja 'High price' avulla
* Visualisoi esim. käyttäen `rolling`-operaattorin antama tasaisesti painotettu sarja
* Visualisoi antaen *enemmän painoa* tuoreimmille havainnoille (`ewm` (exponentially weighted moving) -operaattori).
Tällainen sarja adaptoituu nopeammin muutoksiin kuin `rolling`-operaattorin antama tasaisesti painotettu sarja.


```python

# TODO: Visualisoi data kenttien 'Low price' ja 'High price' avulla

# järjestetään datetime aikajärjestykseen
df.sort_index(inplace=True)

xdata=df.index.tolist()
ydata = df['High price'].tolist()
plt.plot(xdata,ydata,color ='green')

ydata = df['Low price'].tolist()

plt.plot(xdata,ydata,color ='orange')
plt.legend(loc='lower right',labels=['High price','Low price'])

plt.xlabel('Date')
plt.ylabel('Price')

plt.show()

ydata = df['High price'].tolist()

plt.bar(xdata, ydata, color ='green',width = 1)
ydata = df['Low price'].tolist()
plt.bar(xdata, ydata, color ='orange',width = 1)

plt.legend(loc='lower right',labels=['High price','Low price'])
plt.xlabel('Date')
plt.ylabel('Price')

plt.show()

ydata = df['High price'].tolist()
plt.scatter(xdata,ydata, ls='-')
ydata = df['Low price'].tolist()
plt.scatter(xdata,ydata, ls='-')

plt.legend(loc='lower right',labels=['High price','Low price'])
plt.xlabel('Date')
plt.ylabel('Price')

plt.show()
```

Kuvaaja, jossa on mukana Rolling window 20 päivän ajalta.

```python
# TODO: Rolling Window operaattorin käyttö visualisoinnissa

# järjestetään datetime aikajärjestykseen
df.sort_index(inplace=True)

def prog(col):
    xdata=df.index.tolist() 
    
    ydata = df[col].tolist()
    plt.plot(xdata,ydata,color ='blue')

    ydata = df[col].rolling(20).mean().tolist()
    plt.plot(xdata,ydata,color ='orange')

    plt.legend(loc='lower right',labels=['normal','rolling window'])

    plt.xlabel('Date')
    plt.ylabel(col)
    plt.show()

prog('Low price')
prog('High price')
```

#### Tuoreempien havaintojen painottaminen

Usein käytetään menetelmiä, missä annetaan *enemmän painoa* tuoreimmille havainnoille.
Tällainen sarja adaptoituu nopeammin muutoksiin kuin `rolling`-operaattorin antama tasaisesti painotettu sarja.

`ewm` (*exponentially weighted moving*) -operaattori antaa tällaisen sarjan, span-parametri määrittää ikkunan koon.

```python
# TODO: ewn operaattorin käyttö visualisoinnissa

# järjestetään datetime aikajärjestykseen
df.sort_index(inplace=True)

def prog(col):
    xdata=df.index.tolist() 
    
    ydata = df[col].tolist()
    plt.plot(xdata,ydata,color ='blue')
    
    ydata=df[col].ewm(alpha=0.1).mean().tolist()
    plt.plot(xdata,ydata,color ='green')
    plt.legend(loc='lower right',labels=['normal','ewm'])

    plt.xlabel('Date')
    plt.ylabel(col)
    plt.show()

prog('Low price')
prog('High price')
```

## Outlierit ja korrelaatiot (T6.4 / 1p)

Toteuta seuraavat kohdat tässä tehtävässä:

* Datasta tulee tarkastaa ainakin jokaisen numeerisen sarakkeen merkittävät poikkeamat. Merkittävästi poikkeava arvo voidaan tarvittaessa korvata (millä tavalla?).
* Rivin poikkeamien havaitsemiseen voi käyttää `2%` ja `98%` prosenttipisteitä datasta
* laske myös **Pearsonin korrelaatiokerroin** sekä **selityskerroin**.

```python
# TODO: laske poikkeamat (2% - 98%)

df.sort_index(inplace=True)

def prog(column):
    mean=df[column].mean()
    df2=df.loc[df[column] >=mean*1.02]
    df98=df.loc[df[column] <=mean*0.98]
    
    print(column+" >=2% yli keskiarvon")
    print(str(len(df2))+" values")
    print()
    print(df2[column])
    
    print()
    
    ydata = df2[column].tolist()
    xdata = df2.index.tolist()
    plt.scatter(xdata,ydata,color ='blue',s=20)
    plt.axhline(mean,color='r',linestyle='--',label='Avg')
    plt.legend(loc='upper left', frameon=True, markerscale=2)
    plt.xlabel('Date')
    plt.ylabel(column)
    plt.show()
    
    print(column+" <=2% alle keskiarvon")
    print(str(len(df98))+" values")
    print()
    print(df98[column])
    print()
    ydata = df98[column].tolist()
    xdata = df98.index.tolist()
    plt.scatter(xdata,ydata,color ='blue',s=20)
    plt.axhline(mean,color='r',linestyle='--',label='Avg')
    plt.legend(loc='upper left', frameon=True, markerscale=2)
    plt.xlabel('Date')
    plt.ylabel(column)
    plt.show()


for column in df.columns:
    prog(column)    
```

```python
# TODO: Visualisointi regressiomallille
import seaborn as sns

def prog(col1,col2):

    sns.regplot(x = col1, 
                y = col2, 
                data = df,scatter_kws={"color": "orange",'s':10}, line_kws={"color": "black"})
    plt.show()

for col1 in df.columns:
    for col2 in df.columns:
        if col1!=col2:
            prog(col1,col2)
```

#### Regressiomalli -  Lineaarinen malli riippuvuudelle

Lineeaarinen malli riippuvuudelle saadaan tuomalla `sklearn.lineal_model`-kirjastosta `LinearRegression`-luokka.
* valitse **selittävä muuttuja (x)** sekä **selitettävä muuttuja (y)**

```python
from sklearn.linear_model import LinearRegression

# TODO: Regressiomalli: Lineaarinen malli valituille kentille.
# TODO: Perustele myös valinta.

def prog(col1,col2):

    malli=LinearRegression()
    x = df[col1].to_frame()  # muutetaan DataFrameksi
    y = df[col2]
    malli.fit(x,y)
    
    print("kulmakerroin: "+str(malli.coef_))  # kulmakerroin
    
    xfit = df[col1].tolist()
    Xfit = pd.DataFrame(xfit)
    yfit = malli.predict(Xfit)
    plt.scatter(x,y,color='grey',s=10)
    # piirretään "suora"
    plt.plot(xfit,yfit,color='b')
    plt.xlabel(col1)
    plt.ylabel(col2)
    plt.show()

    
#prog('Closing price','High price')
#prog('Closing price','Low price')
prog('Opening price','Closing price')
#prog('High price','Total volume')
#prog('Low price','Total volume')

print("vertasin avauskurssia ja päätöskurssia")
print("kulmakerroin 0.98 on jo lähellä yhtä eli kurssikehitys on hyvin stabiilia päivän aikana ja muutenkin")

'''
for col1 in df.columns:
    for col2 in df.columns:
        if col1!=col2:
            prog(col1,col2)
'''
```
