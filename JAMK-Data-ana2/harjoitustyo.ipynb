```python
import pandas as pd
import requests
url="https://rata.digitraffic.fi/api/v1/live-trains/station/HKI?arrived_trains=5&arriving_trains=10&departed_trains=5&departing_trains=10&include_nonstopping=false"
req = requests.get(url=url)
#print(req.status_code)
#print(req.json())
data=req.json()
```

## Vaihe 2

#### Tehtävä 7. Pura tämän kansion sisällä olevan `data.html` -tiedoston `table`-elementin solujen arvot käyttäen regex-syntaksia ja yhdistä ne vaiheessa 1 luotuun dataframeen. Huom: tehtävässä tulee käyttää vain regexiä (Esim. Pandasin read_html()-metodin käyttö on kielletty)!

```python
import re
import pandas as pd

def readFile(file):
    with open(file, "r") as file:
        content = ""
        for readline in file: 
            line = readline.strip()
            content += line
        return content

def getTable(str):
    return re.search(r'<table>(.*?)</table>', str).group(1)
    
def getHeaders(str):

    startTag=re.search(r'<.*?>', str).group(0)    
    tag=[startTag,'</tr>']
    pattern=tag[0]+"(.*?)"+tag[1]
    str=re.search(pattern, table).group(1)
    headers=[]
    tag=['<th>','</th>']
    pattern=tag[0]+"(.*?)"+tag[1]

    while(str!=""):
        val=re.search(pattern, str).group(1)
        headers.append(val)
        str=str[len(val)+len(tag[0])+len(tag[1]):len(str)]
    return headers

def getRows(str):
    rows=[]
    
    while(str!=""):
        startTag=re.search(r'<.*?>', str).group(0)        
        tag=[startTag,'</tr>']
        pattern=tag[0]+"(.*?)"+tag[1]
        val=re.search(pattern, str).group(1)
        rows.append(val)
        str=str[len(val)+len(tag[0])+len(tag[1]):len(str)]    
    return rows

def getColumns(row):
    cols=[]
    tag=['<td>','</td>']
    pattern=tag[0]+"(.*?)"+tag[1]
    cols=re.findall(pattern, row)
    return cols

str=readFile("data.html")
table=getTable(str)
headers=getHeaders(table)

startTag=re.search(r'<.*?>', table).group(0)    
tag=[startTag,'</tr>']
pattern=tag[0]+"(.*?)"+tag[1]    
headerStr=re.search(pattern, table).group(1)
str=table[len(headerStr)+len(tag[0])+len(tag[1]):len(table)]
rows=getRows(str)

df2 = pd.DataFrame(columns=headers)
for row in rows:
    df2.loc[len(df2.index)] = getColumns(row)   
    
newdf = pd.concat([newdf, df2], axis=0).reset_index(drop=True)

newdf
 
```
